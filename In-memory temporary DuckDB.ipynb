```python
import duckdb
# create an in-memory temporary database by using duckdb.connect()
conn = duckdb.connect()
```

```python
df = pd.read_csv("./AdventureWorksSalesData2020.csv")
```

```python
df.head()
```

```python
conn.register("AdventureWorksSales", df)
```

```python
df_tk_9 = conn.execute("""

SELECT * FROM AdventureWorksSales WHERE TerritoryKey > 9

""").fetchdf()
```
